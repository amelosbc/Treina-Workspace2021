## desafio 1 - pesquisar porque a classe tratamento é tão desbalanceada ?

```python
#preparando dados

import pandas as pd
import matplotlib.pyplot as plot

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = "zip")
dados.columns = dados.columns.str.replace('-','')
dados
```

## desafio 3 - proporção das classes tratamento


```python

##dados['tratamento'].value_counts().plot.pie(shadow=True)

tipos_de_tratamento = dados.tratamento.value_counts()
labels = 'Com drogas', 'Com controle'

plot.title('Tipos de tratamento')
plot.pie(tipos_de_tratamento, labels=labels, labeldistance=1.1)
plot.legend()
plot.show()
```

# desafio 8 - deixar o grafico bonitão, com biblioteca 
# https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

```python
import matplotlib.pyplot as plt

tipo_de_tratamento = dados['tratamento'].value_counts()
labels = dados.tratamento.unique()
explode = (0, 0.2)

plt.title('Tipos de tratamento')
plt.pie(tipo_de_tratamento, labels = labels, shadow=True, explode=explode, autopct='%1.1f%%')
```

```python
tipo_de_doses = dados['dose'].value_counts()
labels = dados.dose.unique()
explode = (0, 0.2)

plt.title('Tipos de doses')
plt.pie(tipo_de_doses, labels = labels, shadow=True, explode=explode, autopct='%1.1f%%')
```

```python
tempos = dados['tempo'].value_counts()
labels = dados.tempo.unique()
#explode = (0, 0.1)

plt.title('Tempos usados')
plt.pie(tempos, labels = labels, shadow=True,  autopct='%1.1f%%')
```
